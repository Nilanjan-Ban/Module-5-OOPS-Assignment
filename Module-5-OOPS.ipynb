```python
'''10.What are class and static methods in Python?

Ans:  In Python, class methods and static methods are used to define behaviors related to a class rather than an instance. They are defined using decorators:

@classmethod → Used to define class methods.
@staticmethod → Used to define static methods.
1. Class Methods (@classmethod)
A class method takes the class (cls) as its first argument and can modify class-level attributes.

Characteristics:
✅ Defined using @classmethod.
✅ Takes cls as the first parameter (instead of self).
✅ Can modify class attributes, but not instance attributes.
✅ Can be called using both class name and object.

Example of Class Method

class Car:
    wheels = 4  # Class attribute (shared by all instances)

    def __init__(self, brand):
        self.brand = brand  # Instance attribute

    @classmethod
    def change_wheel_count(cls, new_count):
        cls.wheels = new_count  # Modifies the class attribute

# Call class method without creating an instance
Car.change_wheel_count(6)
print(Car.wheels)  # Output: 6

# Create instances and check if they reflect the change
car1 = Car("Toyota")
car2 = Car("Honda")

print(car1.wheels)  # Output: 6
print(car2.wheels)  # Output: 6
✅ Class methods modify class attributes, affecting all instances.

2. Static Methods (@staticmethod)
A static method does not take self or cls as its first argument. It behaves like a regular function but is included inside a class for organizational purposes.

Characteristics:
✅ Defined using @staticmethod.
✅ Does not take self or cls as an argument.
✅ Cannot access or modify instance or class attributes.
✅ Used for utility functions that logically belong to a class.

Example of Static Method

class MathOperations:
    @staticmethod
    def add(x, y):
        return x + y  # Simple function inside a class

# Calling static method
print(MathOperations.add(5, 3))  # Output: 8
✅ A static method acts like a normal function but is grouped inside a class for better organization.
'''
```

```python
'''11. What is method overloading in Python?

Ans: In Python, method overloading refers to the ability to define multiple methods with the same name but with different parameter signatures (i.e., a different number or type of parameters). However, Python does not natively support method overloading in the way some other languages like Java or C++ do, where you can define multiple methods with the same name but different parameters.

How Does Python Handle Method Overloading?
In Python, if you define a method with the same name multiple times within the same class, the last definition overrides the previous ones. This is because Python allows only one method with a given name within a class.

However, we can achieve method overloading by:

Using default arguments.
Using variable-length argument lists like *args and **kwargs.
1. Achieving Method Overloading Using Default Arguments
class Calculator:
    def add(self, a, b=0, c=0):  # Default values for b and c
        return a + b + c

calc = Calculator()
print(calc.add(5))        # Output: 5 (only one argument)
print(calc.add(5, 3))     # Output: 8 (two arguments)
print(calc.add(5, 3, 2))  # Output: 10 (three arguments)
✅ Explanation: The method add() has default values for b and c, allowing it to behave like an overloaded method with different numbers of arguments.

2. Achieving Method Overloading Using *args (Variable Arguments)
class Calculator:
    def add(self, *args):  # Accept any number of arguments
        return sum(args)

calc = Calculator()
print(calc.add(1))            # Output: 1
print(calc.add(1, 2, 3))       # Output: 6
print(calc.add(1, 2, 3, 4, 5)) # Output: 15
✅ Explanation: The *args allows the add() method to accept a variable number of arguments, which simulates method overloading.

3. Achieving Method Overloading Using **kwargs (Keyword Arguments)
class Calculator:
    def add(self, **kwargs):
        return sum(kwargs.values())

calc = Calculator()
print(calc.add(a=1))          # Output: 1
print(calc.add(a=1, b=2))     # Output: 3
print(calc.add(a=1, b=2, c=3)) # Output: 6
✅ Explanation: The **kwargs allows you to pass a variable number of named arguments (as a dictionary), simulating method overloading with named parameters.

'''
```

```python
'''22. What is the difference between @staticmethod and @classmethod in Python?

Ans:In Python, both @staticmethod and @classmethod are decorators that are used to define methods within a class that are not bound to an instance of the class (i.e., they do not require access to self for instance methods). However, they serve different purposes and behave in distinct ways:

1. @staticmethod:
Definition: A @staticmethod is a method that does not take a reference to the instance (self) or the class (cls) as
 its first argument. It behaves just like a normal function but belongs to the class's namespace.

Purpose: Use @staticmethod when you need a method that is logically related to the class but does not need to
access or modify any properties or methods of the class or instances of the class.

Access: A static method can be called on the class itself, or on instances of the class, but it cannot access
or modify the instance (self) or class (cls) attributes.

Syntax:
class MyClass:
    @staticmethod
    def my_method(arg1, arg2):
        # method logic here
        print(arg1, arg2)

# Using static method
MyClass.my_method(10, 20)  # Output: 10 20
When to Use @staticmethod:
When the method doesn't need to access or modify instance or class attributes.
When the method is a utility function that logically belongs to the class, but does not interact with the class or instance.
2. @classmethod:
Definition: A @classmethod is a method that takes the class (cls) as its first argument (instead of the instance self). It can access and modify class-level attributes but not instance-level attributes.

Purpose: Use @classmethod when you need a method that operates on the class itself (rather than instances) or when you want to create factory methods (methods that create new instances of the class).

Access: A class method can access and modify the class's state, i.e., it can access class variables and methods.

Syntax:


class MyClass:
    class_variable = 0

    @classmethod
    def class_method(cls, arg1):
        cls.class_variable += arg1
        print(f"Class Variable: {cls.class_variable}")

# Using class method
MyClass.class_method(10)  # Output: Class Variable: 10
When to Use @classmethod:
When we need to operate on or modify class-level data (class variables).
When we want to create factory methods that return instances of the class.
When we want a method that is related to the class itself, rather than instances of the class.

'''
```

```python
'''23. How does polymorphism work in Python with inheritance?

Ans:  Polymorphism is a fundamental concept in object-oriented programming (OOP) that allows objects of different classes to be treated as objects of a common superclass. In Python, polymorphism enables you to use a method in different classes that have the same name, but each class can have its own implementation of the method.

In the context of inheritance, polymorphism means that a subclass can override a method defined in its parent class. When you call that method on an object of the subclass, Python uses the overridden version of the method rather than the version from the parent class.

How Polymorphism Works with Inheritance in Python
Method Overriding:

In polymorphism, the subclass overrides the method of the superclass.
Even if you call the method on an object of the superclass type, the subclass's version of the method gets executed.
Dynamic Method Dispatch:

Python uses dynamic dispatch (or late binding), meaning that the method that gets executed is determined at runtime, not compile-time.
The appropriate method for the actual object type (not the reference type) is called.
Method Signature Consistency:

For polymorphism to work smoothly, the methods in the subclasses and superclass usually share the same signature (i.e., same method name and parameters),
 although Python allows different signatures.

Example of Polymorphism in Python with Inheritance
Let's look at a simple example where polymorphism is used with inheritance.
We'll define a base class Animal and two subclasses Dog and Cat. Both subclasses will have an speak() method, but each will have a different implementation.

class Animal:
    def speak(self):
        print("Animal makes a sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

# Polymorphism in action
def make_animal_speak(animal):
    animal.speak()

# Create instances of Dog and Cat
dog = Dog()
cat = Cat()

# Passing the instances to the same function
make_animal_speak(dog)  # Output: Dog barks
make_animal_speak(cat)  # Output: Cat meows
Explanation:
We have a base class Animal with a method speak().
The subclasses Dog and Cat each override the speak() method to provide their own implementation.
In the function make_animal_speak(), we pass instances of Dog and Cat.
Even though the function expects an Animal object, Python will call the appropriate speak() method for the actual object type, demonstrating polymorphism.
In this case:

When we pass a Dog object, it calls the speak() method in the Dog class.
When we pass a Cat object, it calls the speak() method in the Cat class.
'''
```

```python
#2. Write a program to create an abstract class Shape with a method area(). Derive classes Circle and Rectangle from it and implement the area() method in both.

from abc import ABC, abstractmethod
import math

# Abstract class Shape
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

# Derived class Circle
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * (self.radius ** 2)

# Derived class Rectangle
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

# Create instances of Circle and Rectangle
circle = Circle(5)
rectangle = Rectangle(4, 6)

# Calculate and print the area for each shape
print(f"Area of circle: {circle.area():.2f}")  # Output: Area of circle: 78.54
print(f"Area of rectangle: {rectangle.area()}")  # Output: Area of rectangle: 24
```

```python
#3. Implement a multi-level inheritance scenario where a class Vehicle has an attribute type. Derive a class Car and further derive a class ElectricCar that adds a battery attribute.


# Base class
class Vehicle:
    def __init__(self, vehicle_type):
        self.type = vehicle_type

    def display_info(self):
        print(f"Vehicle Type: {self.type}")

# Derived class from Vehicle
class Car(Vehicle):
    def __init__(self, vehicle_type, brand):
        super().__init__(vehicle_type)
        self.brand = brand

    def display_info(self):
        super().display_info()
        print(f"Car Brand: {self.brand}")

# Further derived class from Car
class ElectricCar(Car):
    def __init__(self, vehicle_type, brand, battery_capacity):
        super().__init__(vehicle_type, brand)
        self.battery = battery_capacity

    def display_info(self):
        super().display_info()
        print(f"Battery Capacity: {self.battery} kWh")

# Example usage
tesla = ElectricCar("Sedan", "Tesla", 75)
tesla.display_info()
```

```python
#5.Write a program to demonstrate encapsulation by creating a class BankAccount with private attributes balance and methods to deposit, withdraw, and check balance.

class BankAccount:
    def __init__(self, initial_balance=0):
        self.__balance = initial_balance  # Private attribute

    # Method to deposit money
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited: ${amount}")
        else:
            print("Deposit amount must be positive.")

    # Method to withdraw money
    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew: ${amount}")
        else:
            print("Insufficient balance or invalid amount.")

    # Method to check balance
    def get_balance(self):
        return f"Current Balance: ${self.__balance}"

# Example usage
account = BankAccount(100)  # Initial balance of $100
account.deposit(50)         # Depositing $50
account.withdraw(30)        # Withdrawing $30
print(account.get_balance())  # Checking balance

# Attempting direct access to private variable (will raise an error)
# print(account.__balance)  # AttributeError: 'BankAccount' object has no attribute '__balance'

```

```python
#7. Create a class MathOperations with a class method add_numbers() to add two numbers and a static method subtract_numbers() to subtract two numbers.

class MathOperations:
    # Class method to add two numbers
    @classmethod
    def add_numbers(cls, a, b):
        return a + b

    # Static method to subtract two numbers
    @staticmethod
    def subtract_numbers(a, b):
        return a - b

# Example usage
sum_result = MathOperations.add_numbers(10, 5)   # Using class method
diff_result = MathOperations.subtract_numbers(10, 5)  # Using static method

print(f"Sum: {sum_result}")       # Output: Sum: 15
print(f"Difference: {diff_result}")  # Output: Difference: 5

```

```python
#9.Write a class Fraction with attributes numerator and denominator. Override the str method to display the fraction as "numerator/denominator"

class Fraction:
    def __init__(self, numerator, denominator):
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")  # Prevent division by zero
        self.numerator = numerator
        self.denominator = denominator

    def __str__(self):
        return f"{self.numerator}/{self.denominator}"  # Override str method

# Example usage
fraction1 = Fraction(3, 4)
fraction2 = Fraction(5, 8)

print(fraction1)  # Output: 3/4
print(fraction2)  # Output: 5/8




```

```python
#10. Demonstrate operator overloading by creating a class Vector and overriding the add method to add two vectors.

class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Overloading the + operator
    def __add__(self, other):
        if isinstance(other, Vector):
            return Vector(self.x + other.x, self.y + other.y)
        raise TypeError("Operand must be an instance of Vector")

    def __str__(self):
        return f"({self.x}, {self.y})"

# Example usage
v1 = Vector(2, 3)
v2 = Vector(4, 5)
v3 = v1 + v2  # Using overloaded + operator

print(v3)  # Output: (6, 8)


```

```python
#12. Implement a class Student with attributes name and grades. Create a method average_grade() to compute the average of the grades.

class Student:
    def __init__(self, name, grades):
        self.name = name
        self.grades = grades  # List of grades

    def average_grade(self):
        if not self.grades:  # Check for empty list to avoid division by zero
            return 0
        return sum(self.grades) / len(self.grades)

    def __str__(self):
        return f"Student: {self.name}, Average Grade: {self.average_grade():.2f}"

# Example usage
student1 = Student("Alice", [85, 90, 78, 92])
student2 = Student("Bob", [88, 76, 95, 89, 80])

print(student1)  # Output: Student: Alice, Average Grade: 86.25
print(student2)  # Output: Student: Bob, Average Grade: 85.6


```

```python
#14.  Create a class Employee with a method calculate_salary() that computes the salary based on hours worked and hourly rate. Create a derived class Manager that adds a bonus to the salary.

# Base class - Employee
class Employee:
    def __init__(self, name, hours_worked, hourly_rate):
        self.name = name
        self.hours_worked = hours_worked
        self.hourly_rate = hourly_rate

    # Method to calculate salary
    def calculate_salary(self):
        return self.hours_worked * self.hourly_rate

# Derived class - Manager
class Manager(Employee):
    def __init__(self, name, hours_worked, hourly_rate, bonus):
        super().__init__(name, hours_worked, hourly_rate)  # Initialize the Employee class
        self.bonus = bonus

    # Overriding the calculate_salary method to include bonus
    def calculate_salary(self):
        base_salary = super().calculate_salary()
        return base_salary + self.bonus

# Example usage
employee = Employee("John Doe", 40, 15)  # 40 hours worked, $15 per hour
manager = Manager("Jane Smith", 40, 20, 500)  # 40 hours worked, $20 per hour + $500 bonus

print(f"{employee.name}'s Salary: ${employee.calculate_salary()}")  # Output: John's Salary: $600
print(f"{manager.name}'s Salary: ${manager.calculate_salary()}")    # Output: Jane's Salary: $900

```

```python
#16.Create a class Animal with an abstract method sound(). Create two derived classes Cow and Sheep that implement the sound() method


from abc import ABC, abstractmethod

# Abstract class Animal
class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

# Derived class Cow
class Cow(Animal):
    def sound(self):
        return "Moo"

# Derived class Sheep
class Sheep(Animal):
    def sound(self):
        return "Baa"

# Example usage
cow = Cow()
sheep = Sheep()

print(f"Cow sound: {cow.sound()}")  # Output: Cow sound: Moo
print(f"Sheep sound: {sheep.sound()}")  # Output: Sheep sound: Baa
```

```python
#17.Create a class Book with attributes title, author, and year_published. Add a method get_book_info() that returns a formatted string with the book's details.


class Book:
    def __init__(self, title, author, year_published):
        self.title = title
        self.author = author
        self.year_published = year_published

    # Method to return formatted book details
    def get_book_info(self):
        return f"'{self.title}' by {self.author}, published in {self.year_published}"

# Example usage
book1 = Book("1984", "George Orwell", 1949)
book2 = Book("To Kill a Mockingbird", "Harper Lee", 1960)

print(book1.get_book_info())  # Output: '1984' by George Orwell, published in 1949
print(book2.get_book_info())  # Output: 'To Kill a Mockingbird' by Harper Lee, published in 1960

```
